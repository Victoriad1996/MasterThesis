```python
def vectorfield(w, t, p):
    """
    Defines the differential equations for the coupled spring-mass system.

    Arguments:
        w :  vector of the state variables:
                  w = [x1,y1,x2,y2]
        t :  time
        p :  vector of the parameters:
                  p = [m1,m2,k1,k2,L1,L2,b1,b2]
    """
    x = []
    y = []
    for i in range(int(len(w)/2)):
        x.append(w[2*i])
        y.append(w[2*i+1])
    m = p['m']
    k = p['k']
    L = p['L']
    b = p['b']
    end_point = sum(L)
    n = len(x)
    f = [y[0],
        (-b[0] * y[0] - k[0] * (x[0] - L[0]) + k[1] * (x[1] - x[0] - L[1])) / m[0]]
    for i in range(1, n-1):
        f += [y[i], (-b[i] * y[i] - k[i] * (x[i] - x[i-1] - L[i]) + k[i+1] * (x[i+1] - x[i] - L[i+1])) / m[i]]
    f += [y[n-1],
         (-b[n-1] * y[n-1] + k[n] * (end_point - x[n-1] - L[n]) - k[n-1] * (x[n-1] - x[n-2] - L[n-1])) / m[n-1]]
    return f
```

```python
n = 5

# Parameter values
# Masses:
m = [1.] * n
# Spring constants
k = [5.] * (n+1)
# Natural lengths
L = [1.] * (n+1)
# Friction coefficients
b = [0.] * n


# Initial conditions
# x are the initial displacements
x = []
for i in range(n+1):
    x.append(i + np.random.uniform(-0.5, 0.5))

#x = list(range(1, n+1))

#y initial velocities
# x are the initial displacements
y = []
for i in range(n+1):
    y.append( np.random.uniform(-0.5, 0.5))
    
# ODE solver parameters
abserr = 1.0e-8
relerr = 1.0e-6
stoptime = 10.0
numpoints = 250

# Create the time samples for the output of the ODE solver.
# I use a large number of points, only because I want to make
# a plot of the solution that looks nice.
t = [stoptime * float(i) / (numpoints - 1) for i in range(numpoints)]

# Pack up the parameters and initial conditions:
p = {'m' : m,
    'k': k,
    'L': L,
    'b' : b}

names = []
w0 = []
for i in range(n):
    w0.append(x[i])
    w0.append(y[i])
    names.append('x'+str(i))
    names.append('y'+str(i))


data = Simulate_springs.simulate(vectorfield, w0, t, p ,abserr, relerr, names, title="springs.csv")
```

```python
Simulate_springs.plot_df(data)
```

```python
figure(1, figsize=(6, 4.5))

xlabel('t')
grid(True)
lw = 1
for col in data:
    if col[0] == 'x':
        plot(data[col], linewidth=lw, label=col)

legend()
title('Mass Displacements for the\nCoupled Spring-Mass System')
savefig('springs.png', dpi=100)
```

```python
n = 5

# Parameter values
# Masses:
m = [1.] * n
# Spring constants
k = [5.] * (n+1)
# Natural lengths
L = [1.] * (n+1)
# Friction coefficients
b = [0.] * n


# Initial conditions
# x are the initial displacements
x = []
for i in range(n+1):
    x.append(i + np.random.uniform(-0.5, 0.5))

#x = list(range(1, n+1))

#y initial velocities
# x are the initial displacements
y = []
for i in range(n+1):
    y.append( np.random.uniform(-0.5, 0.5))
    
# ODE solver parameters
abserr = 1.0e-8
relerr = 1.0e-6
stoptime = 10.0
numpoints = 250

# Create the time samples for the output of the ODE solver.
# I use a large number of points, only because I want to make
# a plot of the solution that looks nice.
t = [stoptime * float(i) / (numpoints - 1) for i in range(numpoints)]

# Pack up the parameters and initial conditions:
p = {'m' : m,
    'k': k,
    'L': L,
    'b' : b}

names = []
w0 = []
for i in range(n):
    w0.append(x[i])
    w0.append(y[i])
    names.append('x'+str(i))
    names.append('y'+str(i))

# Call the ODE solver.
wsol = odeint(vectorfield, w0, t, args=(p,),
              atol=abserr, rtol=relerr)

with open('two_springs.csv', 'w', newline='') as csvfile:
    # Print & save the solution.
    for t1, w1 in zip(t, wsol):
        
        spamwriter = csv.writer(csvfile, delimiter=',',
                                quotechar='|', quoting=csv.QUOTE_MINIMAL)
        spamwriter.writerow(w1)

data = pd.read_csv('two_springs.csv',  names=names, delimiter=',')
data.index = t
```

```python
data = pd.read_csv('two_springs.csv',  names=names, delimiter=',')

figure(1, figsize=(6, 4.5))

xlabel('t')
grid(True)
lw = 1
for col in data:
    if col[0] == 'x':
        plot(data[col], linewidth=lw, label=col)

legend()
title('Mass Displacements for the\nCoupled Spring-Mass System')
savefig('two_springs.png', dpi=100)
```
